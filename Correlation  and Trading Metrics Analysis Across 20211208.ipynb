# Loop for getting the correlations by combinations .. correlation will be calculate for common days for each combinations

```python
i=0

results_18_dynamic= pd.DataFrame()

for outer_loop in iteration_list:
    i+=1
    print(outer_loop)
    
    print(i)
    
    for inner_loop in iteration_list:
        
        #print(inner_loop
        temp = df_18_base_day[df_18_base_day['Unique_FlAG']==outer_loop]
        temp2 = df_18_base_day[df_18_base_day['Unique_FlAG']==inner_loop]
        
        
        temp = temp.reset_index()
        temp2 = temp2.reset_index()
        
        
        
        final_temp = pd.merge(left=temp,
                            right=temp2,
                            left_on=['Day','Month','Year'],
                            right_on=['Day','Month','Year'],
                            how = 'inner',suffixes=('_left', '_right'))


        res = final_temp[['P/L using Dynamic Sizing_right','P/L using Dynamic Sizing_left']].corr()
        res.iloc[0,1]
        
        test = pd.DataFrame()
        
        test.loc[1,'Condition_right'] = temp.loc[0,'Condition']
        test.loc[1,'Strategy_right'] = temp.loc[0,'Strategy']
        test.loc[1,'Currency_right'] = temp.loc[0,'Currency']
        
        
        test.loc[1,'Condition_left'] = temp2.loc[0,'Condition']
        test.loc[1,'Strategy_left'] = temp2.loc[0,'Strategy']
        test.loc[1,'Currency_left'] = temp2.loc[0,'Currency']
        
        
        
        test.loc[1,'Combination_left'] = inner_loop
        
        test.loc[1,'Combination_right'] = outer_loop
        test.loc[1,'Common_days'] = len(final_temp)
        test.loc[1,'Correlation'] = res.iloc[0,1]
        
        results_18_dynamic = results_18_dynamic.append(test)
        
```

### loopp for getting the metrics by combinations

```python
i=0

results= pd.DataFrame()


df=df_18_base_day

iteration_list = df['Unique_FlAG'].unique()

path = "C:/Users/Avinash Mishra/Avinash/Python Codes/WSG Market/Time Series Forecasting/Individual Charts/18 combination/"

for outer_loop in iteration_list:
    
        try:
        
            i+=1
            print(i)
            temp = df[df['Unique_FlAG']==outer_loop]
            temp=temp.reset_index()



            Equity = temp['Profit']/20000    ### converting to 0.5% risk for the final calcualtion

            test = pd.DataFrame()

            test.loc[1,'Combination'] = outer_loop
            test.loc[1,'Condition'] = temp.loc[1,'Condition']
            test.loc[1,'Strategy'] = temp.loc[1,'Strategy']
            test.loc[1,'Currency'] = temp.loc[1,'Currency']

            test.loc[1,'sharp_ratio'] = qs.stats.sharpe(Equity)
            test.loc[1,'VAR'] = qs.stats.value_at_risk(Equity)
            test.loc[1,'value_at_risk'] = qs.stats.value_at_risk(Equity)
            test.loc[1,'avg_return'] = qs.stats.avg_return(Equity)
            test.loc[1,'max_drawdown'] = qs.stats.max_drawdown(Equity)

            #test.loc[1,'CAGR'] = qs.stats.cagr(Equity)

            var = outer_loop.replace('/','_')

            temp = temp.set_index('Date')
            qs.reports.html(temp['P/L using Dynamic Sizing']/20000,title = outer_loop,output=path + var + "_dynamic.html")


            results = results.append(test)
            
            clear_output()
        except :
            print("error at " + outer_loop)
            

results.to_excel(path + "Dynamic_Metrics_by_combination.xlsx")
        
```
